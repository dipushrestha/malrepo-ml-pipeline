# Exploratory Data Analysis

Explore the image dataset before training.

```python
import numpy as np
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
```

```python
data_dir = Path('data/processed/train')
classes = sorted(d.name for d in data_dir.iterdir() if d.is_dir()) if data_dir.exists() else []
print(f'Number of classes: {len(classes)}')
print(f'Classes: {classes[:10]}')
```

```python
# Class distribution
counts = {}
for cls in classes:
    counts[cls] = len(list((data_dir / cls).glob('*')))

plt.figure(figsize=(12, 4))
plt.bar(range(len(counts)), list(counts.values()))
plt.xlabel('Class Index')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.show()
```

```python
# Sample images
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, cls in enumerate(classes[:10]):
    imgs = list((data_dir / cls).glob('*'))
    if imgs:
        img = Image.open(imgs[0])
        axes[i//5][i%5].imshow(img)
        axes[i//5][i%5].set_title(cls)
        axes[i//5][i%5].axis('off')
plt.tight_layout()
plt.show()
```
